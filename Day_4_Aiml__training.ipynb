```python
# @title Sales and Performance Dashboard
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# STEP 1: Create sample sales data
# ------------------------------
# We are creating a dictionary with sample company sales data.
# Each list represents one column of the dataset.

data = {
    "Salesperson": ["Rahul", "Priya", "Aman", "Neha", "Rahul", "Priya", "Aman", "Neha"],
    "Product": ["Laptop", "Mouse", "Keyboard", "Laptop", "Keyboard", "Mouse", "Laptop", "Mouse"],
    "Month": ["January", "January", "January", "January", "February", "February", "February", "February"],
    "Units Sold": [5, 20, 15, 4, 18, 25, 6, 30],
    "Unit Price": [50000, 500, 1000, 50000, 1000, 500, 50000, 500]
}

# Convert dictionary into a DataFrame
df = pd.DataFrame(data)
```

```python
# STEP 2: Create a new column for Total Sales
# ------------------------------
# Total Sales = Units Sold × Unit Price

df["Total Sales"] = df["Units Sold"] * df["Unit Price"]
```

```python
# STEP 3: Show complete dataset
# ------------------------------
print("===== SALES DATASET =====")
print(df)
```

```python
# STEP 4: Basic dashboard statistics
# ------------------------------

# Total overall sales
total_sales = df["Total Sales"].sum()

# Average sales value
average_sales = df["Total Sales"].mean()

# Product-wise total sales
product_sales = df.groupby("Product")["Total Sales"].sum()

# Salesperson-wise total sales
salesperson_sales = df.groupby("Salesperson")["Total Sales"].sum()

# Best selling product based on total sales
best_product = product_sales.idxmax()

# Best salesperson based on total sales
best_salesperson = salesperson_sales.idxmax()
```

```python
# STEP 5: Print dashboard summary
# ------------------------------
print("\n===== DASHBOARD SUMMARY =====")
print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
print("Best Product:", best_product)
print("Best Salesperson:", best_salesperson)

print("\n===== PRODUCT-WISE SALES =====")
print(product_sales)

print("\n===== SALESPERSON-WISE SALES =====")
print(salesperson_sales)
```

```python
# STEP 6: Visualization 1
# Product-wise sales bar chart
# ------------------------------
plt.figure(figsize=(8, 5))
product_sales.plot(kind="bar")
plt.title("Product-wise Total Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# STEP 7: Visualization 2
# Salesperson-wise sales bar chart
# ------------------------------
plt.figure(figsize=(8, 5))
salesperson_sales.plot(kind="bar")
plt.title("Salesperson-wise Total Sales")
plt.xlabel("Salesperson")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# STEP 8: Visualization 3
# Monthly total sales
# ------------------------------
monthly_sales = df.groupby("Month")["Total Sales"].sum()

plt.figure(figsize=(8, 5))
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# STEP 9: Visualization 4
# Pie chart for product sales share
# ------------------------------
plt.figure(figsize=(7, 7))
product_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Product Sales Share")
plt.ylabel("")  # Hides the default y-label
plt.tight_layout()
plt.show()
```

```python
# STEP 10: Top performer details
# ------------------------------
print("\n===== TOP PERFORMER DETAILS =====")
top_person_data = df[df["Salesperson"] == best_salesperson]
print(top_person_data)
```

```python
# STEP 11: Save the final dataset to CSV file
# ------------------------------
# This will create a CSV file that you can download or use later.

df.to_csv("sales_dashboard_data.csv", index=False)
print("\nCSV file saved successfully as 'sales_dashboard_data.csv'")
```
